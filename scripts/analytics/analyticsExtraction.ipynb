```python
from pathlib import Path
import pandas as pd
import json
DATA_DIR = Path().cwd()/"data"
MAPS_DIR = Path().cwd().parent/"for_thesis"/"maps"
```

```python
test_file = DATA_DIR/"Piped"/"Blend"/"analytics"/"run_1.analytics"
```

```python
with open(test_file, "r") as file:
    data = json.load(file)
df = pd.json_normalize(data)
df.columns = [col.replace("env_times.", "env_") for col in df.columns]
df.head()
```

```python
all_data = []

for bmap in MAPS_DIR.iterdir():
    bmap_name = bmap.stem
    map_dir = DATA_DIR/bmap_name

    for method in map_dir.iterdir():
        print(f"Inserting {bmap_name} method {method.stem}")
        analytics_folder = method/"analytics"
        for run in analytics_folder.iterdir():
            with open(run, "r") as file:
                data = json.load(file)
            df = pd.json_normalize(data)
            df.columns = [col.replace("env_times.", "env_") for col in df.columns]
            df.insert(1, "method", method.stem)
            df.insert(2, "run", run.stem)
            df.insert(3, "map", bmap_name)
            all_data.append(df)

whole_df = pd.concat(all_data, ignore_index=True)
whole_df.head()
```

```python
len(whole_df)
whole_df.to_csv("analytics_data.csv", index=False)
```
