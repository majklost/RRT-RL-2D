```python
log_dir = "/home/michal/Documents/Skola/bakalarka/RRT-RL-2D/experiments/RL/logs/cable-last-last/cable-last-last_r2_03-03-21-33-32"
reader = SummaryReader(log_dir)
df = reader.scalars
print(df.tag.unique())
# display(df[(df.tag=="rollout/ep_rew_mean")])
sns.lineplot(data=df[(df.tag=="rollout/ep_rew_mean")],x="step",y="value")
plt.xlabel(r"\textbf{steps trn}")
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.ylabel(r"\textbf{Avg reward}")
# plt.title(r"\textbf{Cable last reward}")

plt.savefig("./plots/reward_per_episode.pdf", bbox_inches='tight')
```
